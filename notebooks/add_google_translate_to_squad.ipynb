```python
"""
Author: Arnab Debnath
This code uses the translated squad datasets for each language where the dataset is modified
to have only 1 qa pair per context. It takes random but globally unique instances from the language files and 
merge them to squad.
"""
```

```python
import os
import json
import random
import copy
```

```python
random.seed(50)
```

```python
SQUAD_FILE = '../Assignment-4/dataset/squad_train-v1.1.json'
squad_dataset = json.load(open(SQUAD_FILE))
```

```python
number_of_instances = 50
```

```python
language_path = '../Assignment-4/dataset/google_translate_single_qa'
language_files = os.listdir(language_path)
language_files = [l for l in language_files if '.json' in l]
```

```python
language_files
```

```python
global_title_paragraph = []
```

```python
for language_file in language_files:
    in_file_path = os.path.join(language_path, language_file)
    print(in_file_path)
    in_file = json.load(open(in_file_path))
    
    for n in range(number_of_instances):
        while True:
            title_index = random.randint(0,len(in_file['data'])-1)
            title = in_file['data'][title_index]['title']
            
            paragraphs = in_file['data'][title_index]['paragraphs']
            paragraph_index = random.randint(0, len(paragraphs)-1)
            paragraph = paragraphs[paragraph_index]

            title_paragraph = (title_index, paragraph_index)        
            if title_paragraph not in global_title_paragraph:
                break
        global_title_paragraph.append(title_paragraph)
        
        d = {}
        d['title'] = title
        d['paragraphs'] = [paragraph]
        squad_dataset['data'].extend([d])
```

```python
out_path = f'../Assignment-4/dataset/merged_squad_google_transltate/set3_50'
out_file_name = os.path.join(out_path, f'squad_plus_uniqueTranslatedSquad_{number_of_instances}.json')
with open(out_file_name, 'w', encoding='utf-8') as outfile:
    json.dump(squad_dataset, outfile)
```
