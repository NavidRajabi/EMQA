```python
import os
import json
import copy
import random
```

```python
random.seed(123)
```

```python
SQUAD_FILE_PATH = '../Assignment-4/dataset/squad_train-v1.1.json'
```

```python
squad_dataset = json.load(open(SQUAD_FILE_PATH))
```

```python
languages = ['ar', 'bn', 'en', 'fi', 'id', 'ko', 'ru', 'sw', 'te']
languages_except_english =  ['ar', 'bn', 'fi', 'id', 'ko', 'ru', 'sw', 'te']
count_context_answer = [0,0,0,0,0,0,0,0,0] 
count_quetion = [0,0,0,0,0,0,0,0,0] 
```

```python
translated_squad_path = '../Assignment-4/dataset/google_translate_train'
```

```python
d = {l:{} for l in languages}
```

```python
for l in languages_except_english:
    language_file_name = f'squad_{l}_google.json'
    language_file_path = os.path.join(translated_squad_path, language_file_name)
    data = json.load(open(language_file_path))['data']
    for i in data:
        paragraphs = i['paragraphs']
        for paragraph in paragraphs:
            context = paragraph['context']
            qas = paragraph['qas']
            for qa in qas:
                answer = qa['answers']
                question = qa['question']
                id_no = qa['id']
                if id_no in d[l]:
                    print("same ID") 
                else:
                    d[l][id_no] = {}
                    d[l][id_no]['context'] = context
                    d[l][id_no]['question'] = question
                    d[l][id_no]['answer'] = answer
```

```python
for i in squad_dataset['data']:
    paragraphs = i['paragraphs']
    for paragraph in paragraphs:
        context = paragraph['context']
        qas = paragraph['qas']
        for qa in qas:
            answer = qa['answers']
            question = qa['question']
            id_no = qa['id']
            if id_no in d['en']:
                print("same ID") 
            else:
                d['en'][id_no] = {}
                d['en'][id_no]['context'] = context
                d['en'][id_no]['question'] = question
                d['en'][id_no]['answer'] = answer
```

```python
#Check if there's a possibility that some ID is in English but not in any of the other languages.
src_language = 'en'
other_languages = [l for l in languages if l != src_language]
only_en_id = []
for i in d[src_language]:
    flag = False
    for l in other_languages:
        if i in d[l]:
            flag = True
            break
    if flag == False:
        only_en_id.append(i)
```

```python
len(only_en_id), len(d['en'])
```

```python
no_of_paragragraph_removed = 1
for index, i in enumerate(squad_dataset['data']):
    title = i['title']
    paragraphs = i['paragraphs']
    k = 0
    while k < len(paragraphs):
        paragraph = paragraphs[k]
        context = paragraph['context']
        
                    
        qas = paragraph['qas']
        j = 0
        flag_context_answer = False
        while j < len(qas):
            qa = qas[j]
            answer = qa['answers'][0] #For each question, there is only 1 answer.
            question = qa['question']
            id_no = qa['id']
            
            if id_no in only_en_id:
                qas.pop(j)
                continue
            
            if flag_context_answer == False:
                language_context_answer_list = ['ar', 'bn', 'en', 'fi', 'id', 'ko', 'ru', 'sw', 'te']
                while True:
                    random_context_answer_language = random.choice(language_context_answer_list)
                    if id_no in d[random_context_answer_language]:
                        flag_context_answer = True
                        break
                    language_context_answer_list.remove(random_context_answer_language)
            
            if id_no not in d[random_context_answer_language]:
                qas.pop(j)
                continue
            language_question_list = [l for l in languages if l!=random_context_answer_language]
            flag = False
            while True:
                random_question_language = random.choice(language_question_list)
                if id_no in d[random_question_language]:
                    flag = True
                    break
                language_question_list.remove(random_question_language)
                if len(language_question_list) == 0:
                    break
            
            if flag == False:
                qas.pop(j)
            else:
                paragraph['context'] = d[random_context_answer_language][id_no]['context']
                qa['answers'] = d[random_context_answer_language][id_no]['answer']
                qa['question'] = d[random_question_language][id_no]['question']
                
                count_context_answer[languages.index(random_context_answer_language)] += 1
                count_quetion[languages.index(random_question_language)] += 1
                j += 1
        if len(qas) == 0:
            print(f"paragraphs removed: {no_of_paragragraph_removed}")
            paragraphs.pop(k)
            no_of_paragragraph_removed += 1
        else:
            k += 1
    if len(paragraphs) == 0:
        print("Every paragraph removed for a title")
    print(index)
```
